```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Simple Linear Regress Model

```
dataset = pd.read_csv("/content/drive/My Drive/MachineLearning/Linear Regression/simple-Linear-Regression-master/Salary_Data.csv")
```

```
dataset.head(100)
```

```
# Two parameters rows, columns for the iloc function

x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

```
x
```

```
y
```

```
## Splitting the dataset

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/3, random_state=0)
```

```
x_train
```

```
y_train
```

```
## Implementatio of classifier based on simple linear regression
```

```
from sklearn.linear_model import LinearRegression

simpleLinearRegression = LinearRegression()
```

```
simpleLinearRegression.fit(x_train,y_train)
```

```

# Predicting the values

y_predict = simpleLinearRegression.predict(x_test)
```

```
y_predict
```

```
### Few examples to predict the salary for few experiences
y_predict_values = simpleLinearRegression.predict([[11]])
```
